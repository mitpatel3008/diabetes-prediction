```python
zero_check_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("Zero values in medical columns (these are actually missing values):\n")
for col in zero_check_columns:
    zero_count = (df[col] == 0).sum()
    zero_percent = (zero_count / len(df)) * 100
    print(f"  {col:25s} : {zero_count:3d} zeros  ({zero_percent:.1f}%)")
```

```python
print("Class Distribution:")
print(df['Outcome'].value_counts())
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```
